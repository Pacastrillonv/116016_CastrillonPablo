```python

"""
 5. Leer un número entero y determinar el número de cincos que están
    consecutivos en un número. Por ejemplo, en 5512355551 hay 6 cincos
    consecutivos. Ahora, en 234555234512 solo hay 3 cincos consecutivos.
"""
#Se importa la función floor de la librería math
from math import floor

x = (input('Digite un número entero: ')) #Asigna valor a la variable "x"
y = int(x)                               #Realiza una especie de copia de "x"

cons = 0        #Se inicializa la variable de cincos consecutivos en cero
suma = 0        #Empieza la suma de todos los cincos consecutivos existentes
while y > 0:    #en el número ingresado
    num = y - (floor(y/10))*10      #Determina el último digito de 'y'
        
    if (num == 5):            #Si el último dígito es 5
        cons = cons + 1       #Aumenta el valor en 1 para los consecutivos
        suma = suma + 5       #Aumenta el valor de suma en 5 


    else:
        if cons == 1:
            suma = suma - 5    #Si segui es igual a 1, suma se reduce en 5
        segui = 0              #Cada vez que num es diferente de cinco, 
                               #"segui" pasa a ser cero   
    y = int(y/10)           #Continúaa el ciclo con  y//10 para mirar
                            #los dígitos siguientes
    
if (segui == 1):
    suma = suma - 5            #Si "segui" es igual a 1, "suma" se reduce en 5
    
contacinco = int(suma/5)       #Arroja el valor de cincos consecutivos

#Se imprimen los resultados que dependen del valor de contacinco, que es el
#valor que al final nos determina si tenemos o no cincos consecutivos
if contacinco > 1:
    print('En el número',x,'hay',contacinco,'cincos consecutivos')
                              
else:
    print('En el número',x,'NO hay cincos consecutivos')
```

```python
"""

11. Escribir un programa que reciba una cadena de texto y reporte:
a) cuántas letras vocales en mayúscula se entraron;
b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
c) cuántos dígitos se entraron,
d) cuántos espacios se entraron;
e) cuántas palabras reservadas se entraron.

"""
#Se importa el módulo keyword del cual se extraerán varias funciones 
import keyword                                          

#Primero se debe digitar una cadena de texto
cadena = input('Digite una cadena de texto: ')  

vocales = 'AEIOUÜ'
tildes = 'áéíóúÁÉÍÓÚ'  

numvocalesmayu = 0                             
numtildes = 0                                  
numdigitos = 0
numespacios = 0                                
numreservadas = 0                              

#Se inician las variables en cero, luego se realiza un ciclo que adentro lleva
#condicionales, si se cumple alguna de las condiciones, las variables
#aumentarán en uno.
 
for i in cadena:
    if i in vocales:
        numvocalesmayu = numvocalesmayu + 1
        
    elif i in tildes:
        numtildes = numtildes + 1
        
    elif i.isdigit():
        numdigitos = numdigitos + 1
        
    elif i == ' ':
        numespacios = numespacios + 1       

texto_list = cadena.split(' ')
#Se convierte cadena a una lista sin los espacios

#La siguiente función nos indica cuales palabras de nuestra cadena hacen parte
#de las palabras clave o palabras reservadas por python
for i in range(len(texto_list)):
    if texto_list[i] in keyword.kwlist:        
        numreservadas = numreservadas + 1
    
#Imprime los resultados de los contadores de cada variable
print(f'En la cadena de texto ingresada hay: ')
print(f'{numvocalesmayu} vocales en mayúscula')
print(f'{numtildes} letras que llevan tilde')
print(f'{numdigitos} dígitos')
print(f'{numespacios} espacios')
print(f'{numreservadas} palabras reservadas')
```

```python
"""
18. Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U y 
    determinar qué fila de U contiene la
    mayor cantidad de unos.
"""
#Se digita el orden de la matriz cuadrada, es solo un dato ya que el número
#de filas es el mismo número de columnas
n=int(input('Digite el orden de la matriz cuadrada: '))
#No se admite un orden negativo, con int nos aseguramos que el orden sea entero
while n<=0:
    n=int(input('No se admite orden, ingrese orden válido: '))   
print('Matriz cuadrada de orden', n)

#Se construye la matriz cuadrada, se inicia una lista vacía y se va llenando
#progresivamente mientras se cumplen los ciclos limitados por el valor de
#filas y columnas de la matriz, el programa indica la ubicación del valor
#que es añadido a la matriz 
matriz=[]
j=0#Columnas
while j<n:
    fila=[]
    i=0#filas
    while i<n:
        while True:
            try:
                num=float(input(f'Ingrese el número {j+1,i+1}=  '))
                break
            except ValueError:
                print('NO DEBE COLOCAR LETRAS')
        i=i+1
        fila.append(num)
    matriz.append(fila)
 
    j= j+1
    
print('\nLa matriz que usted registró es la siguiente:')

#Se imprime la matriz resultante
for fila in matriz:
    print(fila)

#Se crea la matriz triangular asociada U, que es la matriz que toma la 
#diagonal principal de la matriz y los elementos arriba de ella, los otros 
#elementos valen 0
U=matriz
for i in range (n):
    for j in range(n):
        if j>=i:
            (U[i])[j]=(matriz[i])[j] 
        else:
            (U[i])[j]=0
            
print('\nLa siguiente es la matriz triangular superior U')           
for fila in U:
    print(fila)

#Se analiza fila por fila(lista por lista) de la matriz para determinar 
#cuantos unos posee cada una y mostrar la que más unos tenga
print('\nSe presenta la lista de cantidad de unos presentes en cada \
fila de forma ordenada')
unos=[]
for i in range (n):
     conta= U[i].count(1)
     unos.append(conta)
print(unos)

mayor=unos[0]
for g in range(len(unos)):
    if unos[g]>mayor:
        mayor=unos[g]
pos=[]
for g in range(len(unos)):
    if unos[g] == mayor:
        pos.append((g+1))
        
#Se imprime la mayor cantidad de unos y una lista con el número de fila o fias
#que contiene tal cantidad
print('\nLa mayor cantidad de unos es', mayor,' y se encuentra\
en la(s) fila(s)',pos)
```

```python
'''
19.Leer una lista e identifique la sublista más larga dentro de dicha lista que se encuentra ordenada ascendentemente.
Por ejemplo, en la lista [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la [1, 2, 6, 9].

'''
#Iniciamos con una lista vacía que será llenada progresivamente
lista = []

#Se digita el tamaño de la lista
len1 = int(input('Digite la cantidad de elementos de la lista (tamaño): '))

#No se admiten tamaños de lista negativos
if len1 < 0:
    print('No es un valor válido para la longitud de una lista')
                            
while len1 > 0:
        lista.append((int(input('Digite la lista de números: '))))
        len1 = len1 - 1

#Iniciamos otra lista diferente para en ella meter la lista ascendente                                         
listacons = []                                     
for i in range(len(lista)):
    aux = i       
    aux_2 = -1  #Variables auxiliares
                
    for j in range(i, len(lista)-1): #Realiza la organización de las listas
        if lista[j] <= lista[j+1]:  #de forma ascendente
            aux_2 = j+1
        else:
            break
    listacons.append(lista[aux:aux_2+1])
    
lista_mayor = listacons[0]
for i in listacons:                   
    if len(i) >= len(lista_mayor):
        lista_mayor = i   
#El anterior ciclo separa la sublista ordenada ascendentemente con mayor len
#(longitud) y esta será la sublista más larga
    
print('La lista ingresada es: ', lista)

#Se imprime si hay sublistas ordenadas ascendentemente, si es afirmativo
#sube la sublista mas larga

if len(lista_mayor) == 0:
    print('No se halló una lista.')   
else:
    print(f'La sublista más larga es: {lista_mayor}')
```

```python
"""
20. Hacer un programa que lea una matriz y la rote 90 grados en sentido antihorario.
Imprimir la matriz resultante.
"""
#Llamamos el módulo numpy como np 
import numpy as np

#Se digita en principio el tamaño de la matriz en filas y columnas
while True:
    try:
        m = int(input('Ingrese la cantidad de filas: '))
        n = int(input('Ingrese la cantidad de columnas: '))
        break
    except ValueError:
        print('ERROR. Filas y columnas son cantidades enteras')

matriz = []             #Se llena la matriz fila a fila, a medida que se 
j = 0                   #ingrese cada dígito, el programa indicará qué posición
while j < m:            #está ingresando, el programa no permite ingresar 
    fil = []            #letras, por lo que un error pedirá de nuevo que 
    i = 0               #se digite el valor.
    while i < n:
        while True:
            try:
                num = float(input(f'Ingrese un número {j+1,i+1}: '))
                break
            except ValueError:
                print('Recuerde que solo debe ingresar números')
        i += 1
        fil.append(num)
    matriz.append(fil)
    j += 1

#Imprimimos la matriz original digitada que se va a rotar
print('\nLa matriz ingresada fue:')
for fil in matriz:
    print(fil)

#Se convierte en un array    
matriz = np.array(matriz)

#Se crea una matriz vacía que se irá llenando dentro del ciclo 'for' con append
matriz_90 = []
for i in range(n-1,-1,-1):
    matriz_90.append(list(matriz[:,i]))
   
matriz_90 = np.array(matriz_90)

#Se imprime el resultado de la matriz rotada 90 grados

print('\nLa matriz rotada 90 grados es:')
print(matriz_90)
```
