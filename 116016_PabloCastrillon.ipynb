```python
"""
10.Sin utilizar el método .upper() para cadenas, escribir un programa que lea una cadena de texto y la imprima en
mayúscula. Tenga en cuenta que el programa debe aceptar las letras con tilde, la ñ y la ü.
"""
#Primero ingresamos un texto que será almacenado como una lista con el
#comando list
lista1=(list(input('Ingrese un texto cualquiera: ')))

largo= len(lista1) #Longitud de la lista
largoa= largo#Se genera un tipo de copia para conservar el valor de longitud

lista2=[] #Se inicializa una lista vacía que será llenada en el ciclo
while largo>0: 
    x=lista1[largoa-largo]#Se llama a cada elemento de la lista
    posición= ord(x)
    newp=posición-32#Por medio de ASCII, si se le resta 32 a cada elemento
                    #este quedará guardado en letras mayúsculas
    newe=chr(newp) #Se convierte a caracter mediante chr()
    lista2.append(newe)#A la lista vacía se le añade el nuevo elemento por
    largo -=1           #por medio de append
    
union= ' '.join(lista2) #Se convierte la lista modificada en cadena con el
                        #comando join
print('El texto pasado a mayúsculas es ', union)
```

```python
"""
17. Leer una lista y determinar cuál es el segundo elemento más grande de dicha lista. No utilice ningún tipo de algoritmo
de ordenamiento para realizar este ejercicio.
"""
import sys
#Importa la librería sys, que la llamaremos para que dentro de una lista de 
#elementos cualquiera, clsifique cuál es el segundo elemento con mayor tamaño
lista1=[]
#Iniciamos nuevamente con una lista vacía que se va llenando progresivamente
len1=int(input('Digite el tamaño de la lista: '))

while len1<0:#No se admiten tamaños de listas negativos
    len1=int(input('Valor invalido de longitud, intente de nuevo= '))

#Se digita el valor de la lista paso a paso
for i in range (len1):
    lista1.append(input(f'Digite el elemento {i+1}: '))

print('La lista es: ',lista1)

tamaño1=0 #Se inicia una variable tamaño para verificar cuáñ es el elemento 
        #con mayor tamaño de la lista, posteriormente este se borra, y se
        #repite la iteracion, en dicha repetición ya podemos encontrar el 
        #segundo elemento con mayor tamaño
for i in lista1:
    tamaño= sys.getsizeof(i)#Se usa de la librería sys, la función getsizeof
    if tamaño>=tamaño1:     #que nos brinda el tamaño de cada elemento
        tamaño1=tamaño

for i in lista1:
    if sys.getsizeof(i)==tamaño1:
        lista1.remove(i) #Con remove eliminamos el elemento que mayor valor
                        #o size tenga, para en una segunda iteración encontrar
                        #el valor buscado
tamaño2=0
for i in lista1:
    tamaño=sys.getsizeof(i)#Después de remover un elemento buscamos en la 
    if tamaño>tamaño2:      #nueva lista el elemento con mayor tamaño y ese
        tamaño2=tamaño      #es el valor que el ejercicio nos está pidiendo

for i in lista1:
    if sys.getsizeof(i)==tamaño2:

        print('El segundo elemento con mayor tamaño es', i)
        
```

```python
"""
18. Leer una matriz cuadrada A, obtener su matriz triangular superior asociada U y 
    determinar qué fila de U contiene la
    mayor cantidad de unos.
"""
#Se digita el orden de la matriz cuadrada, es solo un dato ya que el número
#de filas es el mismo número de columnas
n=int(input('Digite el orden de la matriz cuadrada: '))
#No se admite un orden negativo, con int nos aseguramos que el orden sea entero
while n<=0:
    n=int(input('No se admite orden, ingrese orden válido: '))   
print('Matriz cuadrada de orden', n)

#Se construye la matriz cuadrada, se inicia una lista vacía y se va llenando
#progresivamente mientras se cumplen los ciclos limitados por el valor de
#filas y columnas de la matriz, el programa indica la ubicación del valor
#que es añadido a la matriz 
matriz=[]
j=0#Columnas
while j<n:
    fila=[]
    i=0#filas
    while i<n:
        while True:
            try:
                num=float(input(f'Ingrese el número {j+1,i+1}=  '))
                break
            except ValueError:
                print('NO DEBE COLOCAR LETRAS')
        i=i+1
        fila.append(num)
    matriz.append(fila)
 
    j= j+1
    
print('\nLa matriz que usted registró es la siguiente:')

#Se imprime la matriz resultante
for fila in matriz:
    print(fila)

#Se crea la matriz triangular asociada U, que es la matriz que toma la 
#diagonal principal de la matriz y los elementos arriba de ella, los otros 
#elementos valen 0
U=matriz
for i in range (n):
    for j in range(n):
        if j>=i:
            (U[i])[j]=(matriz[i])[j] 
        else:
            (U[i])[j]=0
            
print('\nLa siguiente es la matriz triangular superior U')           
for fila in U:
    print(fila)

#Se analiza fila por fila(lista por lista) de la matriz para determinar 
#cuantos unos posee cada una y mostrar la que más unos tenga
print('\nSe presenta la lista de cantidad de unos presentes en cada \
fila de forma ordenada')
unos=[]
for i in range (n):
     conta= U[i].count(1)
     unos.append(conta)
print(unos)

mayor=unos[0]
for g in range(len(unos)):
    if unos[g]>mayor:
        mayor=unos[g]
pos=[]
for g in range(len(unos)):
    if unos[g] == mayor:
        pos.append((g+1))
        
#Se imprime la mayor cantidad de unos y una lista con el número de fila o fias
#que contiene tal cantidad
print('\nLa mayor cantidad de unos es', mayor,' y se encuentra\
en la(s) fila(s)',pos)
```

```python
"""
Leer una lista e identifique la sublista más larga dentro de dicha lista que se encuentra ordenada ascendentemente.
Por ejemplo, en la lista [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la [1, 2, 6, 9].
"""
#Iniciamos con una lista vacía que será llenada progresivamente
lista = []

#Se digita el tamaño de la lista
len1 = int(input('Digite la cantidad de elementos de la lista (tamaño): '))

#No se admiten tamaños de lista negativos
if len1 < 0:
    print('No es un valor válido para la longitud de una lista')
                            
while len1 > 0:
        lista.append((int(input('Digite la lista de números: '))))
        len1 = len1 - 1

#Iniciamos otra lista diferente para en ella meter la lista ascendente                                         
listacons = []                                     
for i in range(len(lista)):
    aux = i       
    aux_2 = -1  #Variables auxiliares
                
    for j in range(i, len(lista)-1): #Realiza la organización de las listas
        if lista[j] <= lista[j+1]:  #de forma ascendente
            aux_2 = j+1
        else:
            break
    listacons.append(lista[aux:aux_2+1])
    
lista_mayor = listacons[0]
for i in listacons:                   
    if len(i) >= len(lista_mayor):
        lista_mayor = i   
#El anterior ciclo separa la sublista ordenada ascendentemente con mayor len
#(longitud) y esta será la sublista más larga
    
print('La lista ingresada es: ', lista)

#Se imprime si hay sublistas ordenadas ascendentemente, si es afirmativo
#sube la sublista mas larga

if len(lista_mayor) == 0:
    print('No se halló una lista.')   
else:
    print(f'La sublista más larga es: {lista_mayor}')
```

```python
"""
20. Hacer un programa que lea una matriz y la rote 90 grados en sentido antihorario.
Imprimir la matriz resultante.
"""
#Llamamos el módulo numpy como np 
import numpy as np

#Se digita en principio el tamaño de la matriz en filas y columnas
while True:
    try:
        m = int(input('Ingrese la cantidad de filas: '))
        n = int(input('Ingrese la cantidad de columnas: '))
        break
    except ValueError:
        print('ERROR. Filas y columnas son cantidades enteras')

matriz = []             #Se llena la matriz fila a fila, a medida que se 
j = 0                   #ingrese cada dígito, el programa indicará qué posición
while j < m:            #está ingresando, el programa no permite ingresar 
    fil = []            #letras, por lo que un error pedirá de nuevo que 
    i = 0               #se digite el valor.
    while i < n:
        while True:
            try:
                num = float(input(f'Ingrese un número {j+1,i+1}: '))
                break
            except ValueError:
                print('Recuerde que solo debe ingresar números')
        i += 1
        fil.append(num)
    matriz.append(fil)
    j += 1

#Imprimimos la matriz original digitada que se va a rotar
print('\nLa matriz ingresada fue:')
for fil in matriz:
    print(fil)

#Se convierte en un array    
matriz = np.array(matriz)

#Se crea una matriz vacía que se irá llenando dentro del ciclo 'for' con append
matriz_90 = []
for i in range(n-1,-1,-1):
    matriz_90.append(list(matriz[:,i]))
   
matriz_90 = np.array(matriz_90)

#Se imprime el resultado de la matriz rotada 90 grados

print('\nLa matriz rotada 90 grados es:')
print(matriz_90)
```
